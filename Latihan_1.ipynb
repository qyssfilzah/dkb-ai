# Meminta nama dan umur pengguna

```python
nama = input("Masukkan nama Anda: ")
umur = input("Masukkan umur Anda: ")

print(f"Halo, {nama}! Umur Anda adalah {umur} tahun.")
```

# Mengira luas segi empat

```python
panjang = float(input("Masukkan panjang segi empat: "))
lebar = float(input("Masukkan lebar segi empat: "))

luas = panjang * lebar

print(f"Luas segi empat adalah {luas}")
```

# Menentukan nombor ganjil atau genap

```python
nombor = int(input("Masukkan nombor: "))

if nombor % 2 == 0:
    print(f"{nombor} adalah nombor genap.")
else:
    print(f"{nombor} adalah nombor ganjil.")
```

# Paparkan nombor 1 hingga 10 menggunakan gelung

```python
for i in range(1, 11):
    print(i)
```

# Menentukan status lulus atau gagal berdasarkan markah

```python
markah = float(input("Masukkan markah Anda: "))

if markah >= 50:
    print("Selamat! Anda lulus.")
else:
    print("Maaf, Anda gagal.")
```
